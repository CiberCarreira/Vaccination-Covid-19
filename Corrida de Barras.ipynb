```python
df = pd.read_csv('country_vaccinations.csv')
df
```

```python
df['country'].unique()
```

```python
df_total = pd.DataFrame()

df_total['País'] = df.country.copy()
df_total['Vacinados'] = df.total_vaccinations.copy()
df_total['Data'] = df.date.copy()
```

```python
df_total.set_index('Data',inplace=True)
```

```python
df_total
```

```python
vac = df_total.groupby(['País',df_total.index]).sum()['Vacinados']
```

```python
pd.DataFrame(vac).unstack().T
```

```python
df_vacinados = pd.DataFrame(vac).unstack().T.droplevel(level=0)
```

```python
df_vacinados.columns
```

```python
maximo = 0
for pais in paises:
    for data in datas:
        if df_vacinados.loc[data][pais] >= maximo:
            maximo = df_vacinados.loc[data][pais]
        else:
            df_vacinados.loc[data][pais] = maximo
    maximo=0
```
